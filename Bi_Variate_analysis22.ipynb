# Scatter Plot
## Numeric vs Numeric
### Scatter plot for Area In sqft vs Price

```python
plt.figure(figsize=(5,5))
plt.scatter(x=bi_analysis["Area In sqft"],y=bi_analysis["Price"],alpha=0.1,color = "#C11C84")
plt.xlabel("Area In sqft")
plt.ylabel("Price")
plt.show()
```

```python
plt.figure(figsize=(5,5),dpi=150)
sns.scatterplot(x=bi_analysis["No_of_bathrooms"],y=bi_analysis["Area In sqft"],alpha=0.05,marker="*",color="black")
plt.xticks(rotation=90)
plt.show()
```

```python
sns.pairplot(bi_analysis)
```

## Numeric VS Category or Category VS Numeric
### Bar graph for Price vs Area

```python
plt.figure(figsize=(5,5))
plt.bar(x=bi_analysis["City"],height=bi_analysis["Price"],color='blue',edgecolor='red') # bar is vertical, barh gives horizontal bar
plt.title("Prices based on City")
plt.xlabel('Area')
plt.ylabel('Price')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.bar(x=bi_analysis["City"],height=bi_analysis["Area In sqft"],color='#f6b092',edgecolor='red') # bar is vertical, barh gives horizontal bar
plt.title("City VS Area in sqft)")
plt.xlabel('City')
plt.ylabel('Area in sqft')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.bar(x=bi_analysis["Price"],height=bi_analysis["Area"],color='blue',edgecolor='red') # bar is vertical, barh gives horizontal bar
plt.title("Price vs City")
plt.xlabel('City')
plt.ylabel('Price')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.bar(x=bi_analysis["Apartment Type"],height=bi_analysis["Price"],color='blue',edgecolor='red') # bar is vertical, barh gives horizontal bar
plt.title("Price vs BHK Type")
plt.xlabel('BHK Type')
plt.ylabel('Price')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=[20,8])
sns.pie(data=bi_analysis, x="City", y="Price")
plt.title("Price vs City")
plt.xlabel('City')
plt.xticks(rotation = 90)
plt.show()
```
