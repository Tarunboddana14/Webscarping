## Count plot for Type

```python
plt.figure(figsize=[10,6])
plt.title('Apartment Type')
sns.countplot(x='Apartment Type', data=uni_analysis)
plt.ylabel('count')
plt.xlabel('Apartment Type')
plt.xticks(rotation=90)
plt.show()
```

### Observations
1. Count of Appartments is 1813 and the count of Independent houses is 1127
2. According to the countplot, it is evident that the number of apartments for sale significantly surpasses number of independent houses for sale.

## Count Plot for Facing(Direction)

```python
uni_analysis["Facing"].value_counts()
```

```python
plt.figure(figsize=[10,6])
plt.title('Directions')
sns.countplot(x='Facing', data=uni_analysis)
plt.ylabel('count')
plt.xlabel('Direction')
plt.xticks(rotation = 90)
plt.yticks([0,5,50,100,200,500,1000,1500,2000])
plt.show()
```

### Observations
1. Andheri East and Ulwe: These areas have the highest number of occurrences, each with 147 counts, indicating a significant number of properties for sale compared to the other areas listed.
2. Gachibowli and Manikonda: Both areas have 98 counts each, making them the second most areas for sale.
3. All the other areas listed have the same count of 49. This uniformity might indicate a consistent pattern in the number of properties available in each area.

## Pie Chart for Facing

```python
plt.pie(data=uni_analysis, x=uni_analysis["Facing"].value_counts().values, 
        labels=uni_analysis["Facing"].value_counts().index, 
        autopct="%0.2f%%",wedgeprops={"edgecolor":"grey"},shadow=True)
plt.title("Directional Distribution of Apartment Facing")
plt.show()
```

### Observations
1. A significant majority of the apartments, totaling 1372(46.67%), have an unknown facing direction, suggesting a high number of properties where the facing direction is not specified or recorded.
2. East-facing apartments are the most common among those with known directions, with 735 counts(25.00%). This indicates a higher availability of east-facing properties.
3. With 294 counts(10.00%), west-facing apartments are the next most common. This shows a notable but lesser availability compared to east-facing apartments.
4. South-facing apartments are the least common among the known directions, with 147 counts(5.00%). This indicates a lower availability for south-facing properties.

## Pie chart for No_of_bathrooms

```python
uni_analysis["No_of_bathrooms"].value_counts()
```

```python
plt.pie(data=uni_analysis, x=uni_analysis["No_of_bathrooms"].value_counts().values, 
        labels=uni_analysis["No_of_bathrooms"].value_counts().index, 
        autopct="%0.2f%%",wedgeprops={"edgecolor":"grey"},shadow=True)
plt.title("No Of Bathrooms")
plt.show()
```

### Observations
1. The majority of the apartments, with 1666 counts(56.67%), have two bathrooms. This suggests that two-bathroom apartments are most available options.
2. Apartments with four bathrooms are the least common, with only 49 counts(1.67%). This indicates a much lower availability for this configuration, likely reflecting a niche market for very large or luxury apartments.
3. The distribution of the number of bathrooms suggests a strong availabilty for apartments with two bathrooms, followed by one-bathroom and three-bathroom configurations. Four-bathroom apartments are rare.
