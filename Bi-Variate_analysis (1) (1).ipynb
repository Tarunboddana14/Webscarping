```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
bi_analysis = pd.read_csv("analysis_data.csv")
```

```python
bi_analysis
```

```python
bi_analysis.info()
```

# Scatter Plot
## Numeric vs Numeric
### Scatter plot for No_of_bathrooms vs Price

```python
plt.figure(figsize=(5,5))
plt.scatter(x=bi_analysis["No_of_bathrooms"],y=bi_analysis["Price"],alpha=0.1,color = "#ff0000")
plt.xlabel("Area In sqft")
plt.ylabel("Price")
plt.show()
```

```python
plt.figure(figsize=(5,5),dpi=150)
sns.scatterplot(x=bi_analysis["Price"],y=bi_analysis["Area In sqft"],alpha=0.05,marker="*",color = "#C11C84")
```

```python
sns.pairplot(bi_analysis)
```

## Numeric VS Category or Category VS Numeric
### Bar graph for Facing vs Price

```python
plt.figure(figsize=(20,10))
plt.bar(x=bi_analysis["Facing"],height=bi_analysis["Price"],color='blue',edgecolor='red') # bar is vertical, barh gives horizontal bar
plt.title("Prices based on Facing")
plt.xlabel('Facing')
plt.ylabel('Price')
plt.xticks(rotation=90)
plt.show()
```

```python
x_axis = bi_analysis["Facing"].unique() 
x_axis
```

```python
y_axis = bi_analysis["Price"].unique()
y_axis
```

```python
plt.figure(figsize=(10,6))
plt.bar(x=bi_analysis["City"],height=bi_analysis["Area In sqft"],color='#f6b092',edgecolor='red') # bar is vertical, barh gives horizontal bar
plt.title("City VS Area in sqft)")
plt.xlabel('City')
plt.ylabel('Area in sqft')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.bar(x=bi_analysis["City"],height=bi_analysis["Price"],color='blue',edgecolor='red') # bar is vertical, barh gives horizontal bar
plt.title("Price vs City")
plt.xlabel('City')
plt.ylabel('Price')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.bar(x=bi_analysis["No_of_bathrooms"],height=bi_analysis["Price"],color='blue',edgecolor='red') # bar is vertical, barh gives horizontal bar
plt.title("No_of_bathrooms vs Price")
plt.xlabel('No_of_bathrooms')
plt.ylabel('Price')
plt.show()
```

# Line Plot
### Categorical vs Numerical
#### Lineplot for Facing vs Price

```python
plt.figure(figsize=(10,6))
sns.lineplot(data=bi_analysis, x="Facing", y="Price")
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(data=bi_analysis, x="City", y="Price")
plt.xticks(rotation = 90)
plt.show()
```

# Violinplot

```python
sns.violinplot(data=bi_analysis,x="City",y="Price")
plt.xticks(rotation = 90)
plt.show()
```

```python
sns.violinplot(data=bi_analysis,x="City", y ="Area In sqft")
plt.xticks(rotation = 90)
plt.show()
```

```python
sns.violinplot(data=bi_analysis,x="Apartment Type", y ="Price")
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(bi_analysis.corr(),annot=True,cmap='coolwarm',vmin=-1)
plt.show()
```
